## объединяю 4 тома в один файл и оставляю только русские буквы

```python
import glob

read_files = glob.glob("*.txt")

with open("result.txt", "wb") as outfile:
    for f in read_files:
        with open(f, "rb") as infile:
            outfile.write(infile.read())
```

```python
with open('result.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
text = re.sub('[^а-я]','', text)
```

```python
RUS = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
POS = {l: i for i, l in enumerate(RUS)}
```

```python
probabilities = np.zeros((len(RUS), len(RUS)))
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUS), columns=(c for c in RUS))
```

```python
s = ['б']
start = np.random.choice(list(RUS))
s.append(start)
length = 6
for i in range(length):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUS), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```

## вероятность встретить строку из 4-х символов, которая начинается "п", а кончается на "р"

```python
def get_letter_and_prob(prev_letter=None):
    if prev_letter is None:
        for s in list(RUS):
            yield s, 1.0 / 32
    else:
        p = POS[prev_letter]
        prob = probabilities[p]
        for s in list(RUS):
            sp = POS[s]
            yield s, prob[sp]
```

```python
s1 = 'п'
s4 = 'р'
for s2, p2 in get_letter_and_prob(s1):
    for s3, p3 in get_letter_and_prob(s2):
        prob = p1 * p2 * p3 * probabilities[POS[s3], POS[s4]]
        if prob > 0:
                print(s1+s2+s3+s4, prob)
            
```
